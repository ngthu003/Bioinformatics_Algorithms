>  __Generate the $d$-Neighborhood Problem__ - `d_Nbhd(Pattern, d)`

*Objective:* Find all the neighbors of a pattern

*  Input: a DNA string *Pattern* and an integer $d$
*  Output: the collection of strings *Neighbors(Pattern, $d$)*

Def: the $d$-neighborhood *Neighbors(Pattern, $d$)* is the set of all $k$-mer whose Hamming distance from *Pattern* does not exceed $d$.

```python
def d_Nbhd(Pattern, d):
    
    # Trivial cases
    if d == 0:
        return Pattern
    if len(Pattern) == 1:
        return ['A', 'C', 'G', 'T']
    
    Nbhd = []
    Suffix_Pattern = Pattern[1:]
    Suffix_Nbhd = d_Nbhd(Suffix_Pattern, d)
    
    for text in Suffix_Nbhd:
        if HammingDistance(Suffix_Pattern, text) < d:
            for x in ['A', 'C', 'G', 'T']:
                if (x + text) not in Nbhd:
                    Nbhd.append(x + text)
        else:
            if (Pattern[0] + text) not in Nbhd:
                Nbhd.append(Pattern[0] + text)
        
    return Nbhd
```

```python
def HammingDistance(String1, String2):
    if len(String1) != len(String2):
        distance = 'Error: Lengths do not match.'
    else:
        distance = 0
        for i in range(0,len(String1)):
            if String1[i] != String2[i]:
                distance += 1
    return distance
```

```python
pattern = 'ACG'
print('Sample test: d-Neighborhood of ' + pattern + ' is ' + str(d_Nbhd(pattern, 1)))
```
