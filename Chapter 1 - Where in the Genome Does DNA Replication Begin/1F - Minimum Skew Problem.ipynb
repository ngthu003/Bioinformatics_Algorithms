>  __Minimum Skew Problem__ - `MinimumSkew(string)`

*Objective:* Find a position in a genome minimizing the skew

*  Input: a DNA string (*Genome*)
*  Output: all integer(s) $i$ minimizing $\textit{Skew}(\textit{Prefix}_i(\textit{string}))$ over all values of $i$

Def: the __skew__ of a DNA string *Genome*, denoted *Skew(Genome)*, is defined to be the difference between the total number of occurences of *'G'* and *'C'* in *Genome*.

```python
def MinimumSkew( string ):
    
    noChar = len(string)
    
    # Initialize count and count array
    count = 0
    count_array = [count]
    
    # Start counting
    for i in range(0, noChar):
        
        if string[i] == 'G':
            count += 1
        elif string[i] == 'C':
            count -= 1
        
        # Append to count array
        count_array.append(count)
    
    # Get minimum difference
    min_count = min(count_array)
    # Get position of those min
    indices = [i for i, x in enumerate(count_array) if x == min_count]
    
    return indices
```

```python
answer = MinimumSkew('CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG')
print('Sample test: ' + str(answer))
```
