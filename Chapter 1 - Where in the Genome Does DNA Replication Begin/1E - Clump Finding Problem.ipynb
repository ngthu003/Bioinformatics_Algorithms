>  __Clump Finding Problem__ - `ClumpFidning(str, k, L, t)`

*Objective:* Find patterns forming clumps in a string

*  Input: a string (*Genome*), and integers $k, L, t$
*  Output: all distinct $k$-mers forming $(L,t)$-clumps in *Genome*

Def: given integers $L$ and $t$, a string *Pattern* forms an __$(L,t)$-clump__ inside a string *Genome* if there is an interval of *Genome* of length $L$, in which Pattern appears at least $t$ times.

```python
def ClumpFinding( string, k, L, t ):
    
    # Initialize lists of clumps
    clump_list = []
    
    # Loop over each 75-unit interval of string
    for u in range(0, len(string)-L+1):
        tmp_string = string[u:u+L]
    
        # Initliaze list of k-mers
        pattern_list = []
    
        # Loop over each k-mer
        for i in range(0, L-k+1 ):

            # Get current k-mer
            pattern = tmp_string[i:i+k]

            # Check for if this is a NEW pattern (ie. not yet appeared previously)
            if pattern not in pattern_list:

                # Update list of checked k-mers
                pattern_list.append(pattern)

                # Count frequency
                count = 1
                for j in range(i+1, L-k+1 ):
                    if pattern == tmp_string[j:j+k]:
                        count += 1

                # Check for if frequent >= t
                if count >= t:
                    if pattern not in clump_list:
                        clump_list.append(pattern)

    return clump_list
```
