>  __Pattern Matching Problem__ - `PatternMatching(Text,Pattern)`

*Objective:* Find all occurences of a pattern in a string

*  Input: strings of *Text* and *Pattern*
*  Output: indices of all starting positions in *Text* where *Pattern* appears as a substring

```python
def PatternMatching( Text, Pattern ):
    t, p = len(Text), len(Pattern)
    indices = []
    for i in range(0,t-p+1):
        if Pattern == Text[i:i+p]:
            indices.append(i)
    return indices
```
