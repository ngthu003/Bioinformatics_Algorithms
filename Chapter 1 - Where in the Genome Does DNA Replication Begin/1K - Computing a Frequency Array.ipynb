>  __Computing a Frequency Array__ - `FrequencyArray(Text,k)`

*Objective:* Generate the frequency array of a DNA string

*  Input: a DNA string *Text* and an integer $k$
*  Output: the frequency array of $k$-mers in *Text*

```python
DNADict = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
def PatternToNumber(Pattern):
    count = 0
    k = len(Pattern)
    for i in range(k):
        char = Pattern[i]
        count += 4**(k-1-i) * DNADict[char]
    return count
```

```python
def FrequencyArray(Text,k):
    FrequencyArray = [0]*(4**k)
    T = len(Text)
    for i in range(0,T-k+1):
        Pattern = Text[i:i+k]
        j = PatternToNumber(Pattern)
        FrequencyArray[j] = FrequencyArray[j] + 1
    return FrequencyArray
```

```python
print('Sample test: The frequency array is ' + str(FrequencyArray('ACGCGGCTCTGAAA',2)))
```

```python
print('Test 1: The frequency array is ' + str(FrequencyArray('AAAAC',2)))
```
